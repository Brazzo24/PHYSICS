```python
import numpy as np
import matplotlib.pyplot as plt

def synthesize_periodic_signal(harmonics, num_points=720):
    """
    Synthesizes a periodic signal from harmonic components.
    
    Parameters:
    harmonics : list of tuples [(order, real, imag), ...]
        List containing harmonic order, real part, and imaginary part of each component.
    num_points : int, optional
        Number of points to evaluate the function over 0 to 720 degrees.
    
    Returns:
    theta : numpy array
        Angular positions from 0 to 720 degrees.
    signal : numpy array
        Synthesized signal values.
    """
    theta = np.linspace(0, 720, num_points)  # Angular positions
    signal = np.zeros_like(theta, dtype=np.float64)
    
    for order, real, imag in harmonics:
        magnitude = np.sqrt(real**2 + imag**2)
        phase = np.arctan2(imag, real)  # Compute phase in radians
        signal += magnitude * np.cos(np.radians(order * theta) + phase)
    
    return theta, signal

def plot_signal(theta, signal):
    plt.figure(figsize=(10, 5))
    plt.plot(theta, signal, label='Synthesized Signal', color='red')
    plt.axhline(0, color='blue', linewidth=1, linestyle='--')
    plt.xlabel('Angle (degrees)')
    plt.ylabel('Amplitude')
    plt.title('Synthesized Periodic Signal')
    plt.legend()
    plt.grid()
    plt.show()

# Example harmonics [(order, real, imag)]
harmonics = [
    (0.5, 10, 0),
    (1, 100, 50),
    (1.5, 20, 0),
    (2, 50, 0),
    (2.5, 5, 0),
    (3, 10, 0)
]

theta, signal = synthesize_periodic_signal(harmonics)
plot_signal(theta, signal)
```
