```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define system parameters
m_s = 111.8  # Sprung mass (kg)
m_u = 19.1   # Unsprung mass (kg)
k_s = 9500  # Suspension stiffness (N/m)
c_s = 625  # Suspension damping (Ns/m)
k_t = 20000  # Tire stiffness (N/m)
c_t = 3       # Tire damping (commonly neglected)

# Road excitation parameters
A = 0.05  # Amplitude (m)
f = 1.5   # Frequency (Hz)
omega = 2 * np.pi * f  # Angular frequency (rad/s)

# Define equations of motion
def quarter_car(t, X):
    x_s, v_s, x_u, v_u = X  # Unpack state variables
    y_r = A * np.sin(omega * t)  # Road excitation
    dy_r = A * omega * np.cos(omega * t)  # Derivative of road excitation

    # Equations of motion
    dx_s = v_s
    dv_s = (-c_s * (v_s - v_u) - k_s * (x_s - x_u)) / m_s
    dx_u = v_u
    dv_u = (c_s * (v_s - v_u) + k_s * (x_s - x_u) - k_t * (x_u - y_r)) / m_u

    return [dx_s, dv_s, dx_u, dv_u]

# Time span
t_span = (0, 5)  # Simulate for 5 seconds
t_eval = np.linspace(t_span[0], t_span[1], 1000)  # Time points for evaluation

# Initial conditions (rest position)
X0 = [0, 0, 0, 0]

# Solve the system
sol = solve_ivp(quarter_car, t_span, X0, t_eval=t_eval, method='RK45')

# Extract results
t = sol.t
x_s, v_s, x_u, v_u = sol.y

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(t, x_s, label='Sprung Mass Displacement (m)')
plt.plot(t, x_u, label='Unsprung Mass Displacement (m)', linestyle='dashed')
plt.plot(t, A * np.sin(omega * t), label='Road Excitation (m)', linestyle='dotted')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.title('Quarter-Car Model Response to Sine Wave Road Excitation')
plt.legend()
plt.grid()
plt.show()
```

# Maxwell Branch

```python
import scipy.signal as signal

# Define frequency range for analysis
freqs = np.logspace(-1, 2, 1000)  # Log-spaced frequencies from 0.1 Hz to 100 Hz
omega_vals = 2 * np.pi * freqs  # Convert to rad/s

# Compute the transfer function H(j) = X_s(j) / Y_r(j)
# Using mechanical impedance analogy

# Mechanical impedances
Z_m_s = 1j * omega_vals * m_s  # Mass impedance for sprung mass
Z_m_u = 1j * omega_vals * m_u  # Mass impedance for unsprung mass
Z_c = c_s  # Damping impedance
Z_k_s = 1 / (1j * omega_vals * k_s)  # Suspension stiffness impedance
Z_k_t = 1 / (1j * omega_vals * k_t)  # Tire stiffness impedance

# Equivalent impedance of the system
Z_eq = (Z_m_u + Z_k_t) * (Z_c + Z_k_s) / (Z_m_u + Z_k_t + Z_c + Z_k_s)

# Transfer function H(j) = X_s(j) / Y_r(j)
H_omega = Z_eq / (Z_m_s + Z_eq)

# Compute magnitude and phase
magnitude = np.abs(H_omega)
phase = np.angle(H_omega, deg=True)  # Phase in degrees

# Plot Bode magnitude and phase
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.loglog(freqs, magnitude)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude |H(j)|")
plt.title("Frequency Response - Magnitude")
plt.grid(True, which="both")

plt.subplot(2, 1, 2)
plt.semilogx(freqs, phase)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (degrees)")
plt.title("Frequency Response - Phase")
plt.grid(True, which="both")

plt.tight_layout()
plt.show()
```

# With Forces and Accleration

```python
# Re-import necessary libraries after execution state reset
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
from scipy.integrate import solve_ivp

# Road excitation parameters
A = 0.05  # Amplitude (m)
f = 1.5   # Frequency (Hz)
omega = 2 * np.pi * f  # Angular frequency (rad/s)

# Define equations of motion for time-domain analysis
def quarter_car(t, X):
    x_s, v_s, x_u, v_u = X  # Unpack state variables
    y_r = A * np.sin(omega * t)  # Road excitation

    # Equations of motion
    dx_s = v_s
    dv_s = (-c_s * (v_s - v_u) - k_s * (x_s - x_u)) / m_s
    dx_u = v_u
    dv_u = (c_s * (v_s - v_u) + k_s * (x_s - x_u) - k_t * (x_u - y_r)) / m_u

    return [dx_s, dv_s, dx_u, dv_u]

# Time span
t_span = (0, 5)  # Simulate for 5 seconds
t_eval = np.linspace(t_span[0], t_span[1], 1000)  # Time points for evaluation

# Initial conditions (rest position)
X0 = [0, 0, 0, 0]

# Solve the system
sol = solve_ivp(quarter_car, t_span, X0, t_eval=t_eval, method='RK45')

# Extract results
t = sol.t
x_s, v_s, x_u, v_u = sol.y

# Compute accelerations (second derivative of displacement)
a_s = np.gradient(v_s, t)  # Sprung mass acceleration
a_u = np.gradient(v_u, t)  # Unsprung mass acceleration

# Compute forces
F_s = k_s * (x_s - x_u) + c_s * (v_s - v_u)  # Suspension force
F_t = k_t * (x_u - A * np.sin(omega * t))  # Tire force

# --- Frequency-Domain Analysis ---

# Define frequency range
freqs = np.logspace(-1, 2, 1000)  # Log-spaced frequencies from 0.1 Hz to 100 Hz
omega_vals = 2 * np.pi * freqs  # Convert to rad/s

# Mechanical impedances
Z_m_s = 1j * omega_vals * m_s  # Mass impedance for sprung mass
Z_m_u = 1j * omega_vals * m_u  # Mass impedance for unsprung mass
Z_c = c_s  # Damping impedance
Z_k_s = 1 / (1j * omega_vals * k_s)  # Suspension stiffness impedance
Z_k_t = 1 / (1j * omega_vals * k_t)  # Tire stiffness impedance

# Equivalent impedance of the system
Z_eq = (Z_m_u + Z_k_t) * (Z_c + Z_k_s) / (Z_m_u + Z_k_t + Z_c + Z_k_s)

# Transfer function H(j) = X_s(j) / Y_r(j)
H_omega = Z_eq / (Z_m_s + Z_eq)

# Compute accelerations in frequency domain
A_s_freq = -omega_vals**2 * H_omega  # Sprung mass acceleration

# Compute suspension and tire forces in frequency domain
F_s_freq = k_s * (H_omega - Z_eq / (Z_m_s + Z_eq)) + 1j * omega_vals * c_s * (H_omega - Z_eq / (Z_m_s + Z_eq))
F_t_freq = k_t * (Z_eq / (Z_m_s + Z_eq) - 1)  # Tire force

# --- Plot Results ---

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Time-domain forces
axs[0, 0].plot(t, F_s, label="Suspension Force (N)")
axs[0, 0].plot(t, F_t, label="Tire Force (N)", linestyle="dashed")
axs[0, 0].set_xlabel("Time (s)")
axs[0, 0].set_ylabel("Force (N)")
axs[0, 0].set_title("Forces Over Time")
axs[0, 0].legend()
axs[0, 0].grid()

# Time-domain accelerations
axs[0, 1].plot(t, a_s, label="Sprung Mass Acceleration (m/s)")
axs[0, 1].plot(t, a_u, label="Unsprung Mass Acceleration (m/s)", linestyle="dashed")
axs[0, 1].set_xlabel("Time (s)")
axs[0, 1].set_ylabel("Acceleration (m/s)")
axs[0, 1].set_title("Accelerations Over Time")
axs[0, 1].legend()
axs[0, 1].grid()

# Frequency-domain forces (Magnitude)
axs[1, 0].loglog(freqs, np.abs(F_s_freq), label="Suspension Force (N)")
axs[1, 0].loglog(freqs, np.abs(F_t_freq), label="Tire Force (N)", linestyle="dashed")
axs[1, 0].set_xlabel("Frequency (Hz)")
axs[1, 0].set_ylabel("|Force| (N)")
axs[1, 0].set_title("Forces in Frequency Domain")
axs[1, 0].legend()
axs[1, 0].grid(which="both")

# Frequency-domain acceleration (Magnitude)
axs[1, 1].loglog(freqs, np.abs(A_s_freq), label="Sprung Mass Acceleration (m/s)")
axs[1, 1].set_xlabel("Frequency (Hz)")
axs[1, 1].set_ylabel("|Acceleration| (m/s)")
axs[1, 1].set_title("Acceleration in Frequency Domain")
axs[1, 1].legend()
axs[1, 1].grid(which="both")

plt.tight_layout()
plt.show()
```

# Phase


```python
# Compute phase response for forces and accelerations in frequency domain
phase_F_s = np.angle(F_s_freq, deg=True)  # Phase of suspension force
phase_F_t = np.angle(F_t_freq, deg=True)  # Phase of tire force
phase_A_s = np.angle(A_s_freq, deg=True)  # Phase of sprung mass acceleration

# --- Plot Phase Responses ---
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Phase response for forces
axs[0].semilogx(freqs, phase_F_s, label="Suspension Force Phase (degrees)")
axs[0].semilogx(freqs, phase_F_t, label="Tire Force Phase (degrees)", linestyle="dashed")
axs[0].set_xlabel("Frequency (Hz)")
axs[0].set_ylabel("Phase (degrees)")
axs[0].set_title("Phase Response of Forces")
axs[0].legend()
axs[0].grid(which="both")

# Phase response for acceleration
axs[1].semilogx(freqs, phase_A_s, label="Sprung Mass Acceleration Phase (degrees)")
axs[1].set_xlabel("Frequency (Hz)")
axs[1].set_ylabel("Phase (degrees)")
axs[1].set_title("Phase Response of Sprung Mass Acceleration")
axs[1].legend()
axs[1].grid(which="both")

plt.tight_layout()
plt.show()
```
