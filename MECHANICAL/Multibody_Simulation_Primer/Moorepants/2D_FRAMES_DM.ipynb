```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def orientation(theta):
    M_orient = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
    
    return M_orient
```

```python
theta = np.radians(45)
```

```python
Rot = orientation(theta)
Rot
```
