```python
import sympy as sm
import pydy
```

```python
theta = sm.symbols('theta')
A_C_N = sm.Matrix([[sm.cos(theta), sm.sin(theta), 0],
[-sm.sin(theta), sm.cos(theta), 0],
[0, 0, 1]])
```

```python
A_C_N
```

This matrix uniquely describes the orientation between the two reference frames and so we give it its own variable A_C_N.
This matrix ACN maps vectors expressed in the N frame to vectors expressed in the A frame. This matrix has an
important property, which we will demonstrate with SymPy. Start by creating the matrix:

```python
sm.trigsimp(A_C_N.inv())
```

```python
A_C_N.transpose()
```

Notably, the inverse and the transpose are the same here. This indicates that this matrix is a special orthogonal matrix.
All matrices that describe the orientation between reference frames are orthogonal matrices. Following the notation
convention, this holds:

```python
alpha = sm.symbols('alpha')
B_C_A = sm.Matrix([[sm.cos(alpha), sm.sin(alpha), 0],
[-sm.sin(alpha), sm.cos(alpha), 0],
[0, 0, 1]])
B_C_A
```

```python
B_C_N = B_C_A*A_C_N
B_C_N
```

```python
# simplify the expression
sm.trigsimp(B_C_N)
```

Now work with simpy.mechanics instead of "only" sympy

```python
import sympy.physics.mechanics as me
class ReferenceFrame(me.ReferenceFrame):
    def __init__(self,*args, **kwargs):
        kwargs.pop('latexs'
        , None)
        lab = args[0].lower()
        tex = r'\hat{{{}}}_{}'
        super(ReferenceFrame, self).__init__(*args,
        latexs=(tex.format(lab, 'x'),
        tex.format(lab, 'y'),
        tex.format(lab, 'z')),
        **kwargs)
me.ReferenceFrame = ReferenceFrame
```

```python
N = me.ReferenceFrame('N')
N.x, N.y, N.z
```

```python
A = me.ReferenceFrame('A')
B = me.ReferenceFrame('B')
N, A, B
```
