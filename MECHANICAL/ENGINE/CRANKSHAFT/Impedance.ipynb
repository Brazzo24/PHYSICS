```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Define function for computing and plotting mobility
def compute_mobility(m1=1.0, m2=2.0, k1=1000, k2=1500, c1=10, c2=20):
    # Define frequency range (in Hz) and convert to rad/s
    freqs = np.logspace(0, 3, 500)  # 1 Hz to 1000 Hz
    omega = 2 * np.pi * freqs  # Convert to rad/s
    s = 1j * omega  # Laplace variable for frequency response

    # Define system matrices
    M = np.array([[m1, 0], [0, m2]])
    C = np.array([[c1 + c2, -c2], [-c2, c2]])
    K = np.array([[k1 + k2, -k2], [-k2, k2]])

    # Compute impedance and mobility matrices at each frequency
    Z_matrices = np.array([M * (s_i**2) + C * s_i + K for s_i in s])
    Y_matrices = np.linalg.inv(Z_matrices)  # Inverse of impedance = mobility

    # Extract mobility terms (Y11 and Y22 for self-mobility)
    Y11 = Y_matrices[:, 0, 0]
    Y22 = Y_matrices[:, 1, 1]

    # Plot mobility magnitude response
    plt.figure(figsize=(8, 6))
    plt.loglog(freqs, np.abs(Y11), label=r"$Y_{11}$ (Mobility of Mass 1)")
    plt.loglog(freqs, np.abs(Y22), label=r"$Y_{22}$ (Mobility of Mass 2)")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Mobility Magnitude")
    plt.title("Mobility Functions of the 2-DOF System")
    plt.legend()
    plt.grid(which="both", linestyle="--")
    plt.show()

# Create interactive sliders
interact(
    compute_mobility,
    m1=widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description="Mass 1 (kg)"),
    m2=widgets.FloatSlider(value=2.0, min=0.1, max=5.0, step=0.1, description="Mass 2 (kg)"),
    k1=widgets.IntSlider(value=1000, min=100, max=5000, step=100, description="Stiffness k1 (N/m)"),
    k2=widgets.IntSlider(value=1500, min=100, max=5000, step=100, description="Stiffness k2 (N/m)"),
    c1=widgets.IntSlider(value=10, min=0, max=100, step=5, description="Damping c1 (Ns/m)"),
    c2=widgets.IntSlider(value=20, min=0, max=100, step=5, description="Damping c2 (Ns/m)"),
);
```

```python
import numpy as np
import matplotlib.pyplot as plt

# System parameters
m1, m2 = 1.0, 2.0  # Masses
k1, k2 = 1000, 1500  # Stiffness
c1, c2 = 10, 20  # Damping coefficients

# Define frequency range (in Hz) and convert to rad/s
freqs = np.logspace(0, 3, 500)  # 1 Hz to 1000 Hz
omega = 2 * np.pi * freqs  # Convert to rad/s
s = 1j * omega  # Laplace variable for frequency response

# Define system matrices
M = np.array([[m1, 0], [0, m2]])
C = np.array([[c1 + c2, -c2], [-c2, c2]])
K = np.array([[k1 + k2, -k2], [-k2, k2]])

# Compute impedance matrices for each frequency
Z_matrices = np.array([M * (s_i**2) + C * s_i + K for s_i in s])

# Extract impedance matrix elements
Z11 = Z_matrices[:, 0, 0]
Z12 = Z_matrices[:, 0, 1]
Z21 = Z_matrices[:, 1, 0]
Z22 = Z_matrices[:, 1, 1]

# Compute ThÃ©venin equivalent impedance at mass 1
Z_th = Z11 - (Z12 * Z21) / Z22

# Compute ThÃ©venin equivalent force (assuming F2 = 0)
F1_applied = 1.0  # Assume an applied force at mass 1
F_th = F1_applied - Z12 * (0 / Z22)  # F2 = 0 in this case

# Plot ThÃ©venin equivalent impedance magnitude
plt.figure(figsize=(8, 6))
plt.loglog(freqs, np.abs(Z_th), label=r"$|Z_{\text{th}}|$ (ThÃ©venin Impedance)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Impedance Magnitude")
plt.title("ThÃ©venin Equivalent Impedance of the System")
plt.legend()
plt.grid(which="both", linestyle="--")
plt.show()

print(f"ThÃ©venin Equivalent Force: {F_th} N")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh

# System parameters (Engine-Chassis Model)
m1, m2 = 100, 50   # kg (engine and intermediate support masses)
c1, c2 = 5, 3 # Ns/m (damping coefficients)
k1, k2 = 100000, 500000  # N/m (stiffness values)

# Frequency range
frequencies = np.linspace(0.1, 50, 500)  # Avoid zero to prevent division by zero
omega = 2 * np.pi * frequencies  # Convert to angular frequency

# Mass matrix
M = np.array([[m1, 0], [0, m2]])

# Damping matrix
C = np.array([[c1 + c2, -c2], [-c2, c2]])

# Stiffness matrix
K = np.array([[k1 + k2, -k2], [-k2, k2]])

# Compute natural frequencies and mode shapes
eigenvalues, eigenvectors = eigh(K, M)
natural_frequencies = np.sqrt(eigenvalues) / (2 * np.pi)
print("Natural Frequencies (Hz):", natural_frequencies)
print("Mode Shapes:")
print(eigenvectors)

# Compute impedance matrix at each frequency
Z = K + 1j * omega[:, None, None] * C - (omega[:, None, None] ** 2) * M

# Compute mobility matrix (Y = Z^(-1))
Y = np.linalg.inv(Z)

# Define external harmonic force (acting on engine, mass 1) as frequency-dependent
F1 = 500 * np.ones_like(frequencies)  # Force applied at all frequencies
F2 = 0 * np.ones_like(frequencies)    # No direct force on mass 2
F = np.array([F1, F2]).T  # Transpose to match frequency shape

# Compute system response (X = Y * F) for each frequency
X = np.einsum('ijk,ik->ij', Y, F)  # Adjusted for frequency-dependent force
X1 = X[:, 0]  # Displacement of engine (mass 1)
X2 = X[:, 1]  # Displacement of chassis (mass 2)

# Compute Transmissibility (Vibration Transmission to Chassis)
transmissibility = np.abs(X2 / X1)

# Plot Transmissibility
plt.figure(figsize=(12, 5))
plt.plot(frequencies, transmissibility, label='Transmissibility |X2/X1|')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Transmissibility')
plt.title('Vibration Transmissibility to Chassis')
plt.legend()
plt.grid()
plt.show()

# Plot Forced Response - Magnitude
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(frequencies, np.abs(X1), label='|X1| (Engine)')
plt.plot(frequencies, np.abs(X2), label='|X2| (Chassis)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Displacement Magnitude (m)')
plt.title('Forced Response - Magnitude')
plt.legend()
plt.grid()

# Plot Forced Response - Phase
plt.subplot(1, 2, 2)
plt.plot(frequencies, np.angle(X1, deg=True), label='âˆ X1 (Engine)')
plt.plot(frequencies, np.angle(X2, deg=True), label='âˆ X2 (Chassis)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (degrees)')
plt.title('Forced Response - Phase')
plt.legend()
plt.grid()

plt.show()
```
